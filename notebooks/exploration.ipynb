```python
import plotly.express as px

df['DernierAchat'] = pd.to_datetime(df['DernierAchat'])
# Agrégation par jour ou par mois pour la clarté
df_daily = df.groupby('DernierAchat')['TotalDepense'].sum().reset_index()

fig1 = px.line(df_daily, x='DernierAchat', y='TotalDepense', 
              title='Tendance des Ventes (Chiffre d\'Affaires par Jour)',
              labels={'DernierAchat': 'Date d\'Achat', 'TotalDepense': 'Total ($)'},
              markers=True)
fig1.show()
```
